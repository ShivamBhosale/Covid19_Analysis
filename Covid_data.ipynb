```python
df=pd.read_csv('Covid_Confirmed.csv')
```

```python
df1=pd.read_csv('Covid_death.csv')
```

```python
df2=pd.read_csv('Covid_recovered.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
df=df.drop(["Province/State","Lat","Long"],axis=1)
```

```python
df.head()
```

```python
df=df.groupby(df["Country/Region"]).aggregate('sum')
```

```python
df2=df2.drop(["Province/State","Lat","Long"],axis=1)
df1=df1.drop(["Province/State","Lat","Long"],axis=1)
```

```python
df1=df1.groupby(df1["Country/Region"]).aggregate('sum')
df2=df2.groupby(df2["Country/Region"]).aggregate('sum')
```

```python
print(df2)
```

```python
df=df.T
df1=df1.T
df2=df2.T
```

```python
df.head()
```

```python
df1.head()
```

```python
df2.head()
```

```python
new_cases=df.copy()
for day in range(1, len(df)):
    new_cases.iloc[day]=df.iloc[day]-df.iloc[day-1]
print(new_cases.tail(10))
```

```python
growth_rate=df.copy()
for day in range(1,len(df)):
    growth_rate.iloc[day]= (new_cases.iloc[day]/df.iloc[day-1]) * 100
growth_rate
```

```python
active_cases=df.copy()
for day in range(1,len(df)):
    active_cases.iloc[day]= df.iloc[day]= df2.iloc[day]
active_cases
```

```python
overall_growth_rate =df.copy()

for day in range(1,len(df)):
    overall_growth_rate.iloc[day]=((active_cases.iloc[day] - active_cases.iloc[day-1]) /active_cases.iloc[day-1])*100
```

```python
overall_growth_rate['India'].tail(10)
```

```python
death_rate = df.copy()
for day in range(1,len(df)):
    death_rate.iloc[day]=(df1.iloc[day]/df.iloc[day])*100
```

```python
death_rate["India"].tail(10)
```

```python
hospitalization_rate_estimate = 0.05
hostpitalization_needed=df.copy()
for day in range(1,len(df)):
    hostpitalization_needed.iloc[day]= active_cases.iloc[day] * hospitalization_rate_estimate

hostpitalization_needed["India"].tail()  
        
```

# Visualization

```python
countries=["India","Russia","US","United Kingdom","Brazil","Canada"]
ax=plt.subplot()
ax.set_facecolor('black')
ax.figure.set_facecolor('#121212')
ax.tick_params(axis='x',colors='white')
ax.tick_params(axis='y',colors='white')
ax.set_title('Covid-19 Total Confirmed Cases by cases',color='white')
for country in countries:
    df[country][100:].plot(label=country,figsize=(10,8))
plt.legend(loc='upper left')
plt.show()
```

```python
coun=["India"]
ax=plt.subplot()
ax.set_facecolor('black')
ax.figure.set_facecolor('#121212')
ax.tick_params(axis='x',colors='white')
ax.tick_params(axis='y',colors='white')
ax.set_title('Covid-19 - Confirmed Cases growth rate (Coun)',color='white')
for coun in countries:
    growth_rate[country][200:].plot.bar(figsize=(100,80))
plt.show()

```
